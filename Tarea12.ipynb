```python
#TAREA 1.2
#CARLOS OSEGUEDA
#20212030669

import pandas as np

df_ventas = np.read_csv('./ventas.csv')
df_clima = np.read_csv('./clima.csv')
df_calificaciones = np.read_csv('./calificaciones.csv')
df_calificaciones.head()
```

```python
# Calcular la cantidad total de productos vendidos por categoría
total_por_categoria = df_ventas.groupby('Producto')['Cantidad'].sum()
print(total_por_categoria)

# Encontrar el producto con el mayor total de ventas
producto_mayor_ventas = total_por_categoria.idxmax()
print(f"El producto con el mayor total de ventas es: {producto_mayor_ventas}")

# Calcular el precio promedio de los productos vendidos
precio_promedio = df_ventas['Precio_Unitario'].mean()
print(f"El precio promedio de los productos vendidos es: {precio_promedio:.2f}")
```

```python
# Calcular la temperatura promedio de cada ciudad
temperatura_promedio_ciudad = df_clima.groupby('Ciudad')['Temperatura'].mean()
print(temperatura_promedio_ciudad)

# Encontrar el registro con la temperatura más alta
temp_max = df_clima['Temperatura'].max()
registro_temp_max = df_clima[df_clima['Temperatura'] == temp_max]

# Encontrar el registro con la temperatura más baja
temp_min = df_clima['Temperatura'].min()
registro_temp_min = df_clima[df_clima['Temperatura'] == temp_min]

print(f"Registro con la temperatura más alta:\n{registro_temp_max}")
print(f"Registro con la temperatura más baja:\n{registro_temp_min}")

# Ciudad con la temperatura más alta
ciudad_temp_max = registro_temp_max['Ciudad'].values[0]

# Ciudad con la temperatura más baja
ciudad_temp_min = registro_temp_min['Ciudad'].values[0]

print(f"Ciudad con la temperatura más alta: {ciudad_temp_max}")
print(f"Ciudad con la temperatura más baja: {ciudad_temp_min}")

# Contar registros con temperatura mayor a 30°C
registros_mayor_30 = df_clima[df_clima['Temperatura'] > 30].shape[0]
print(f"Registros con temperatura mayor a 30°C: {registros_mayor_30}")

# Contar días registrados por cada ciudad
dias_por_ciudad = df_clima['Ciudad'].value_counts()
print(dias_por_ciudad)
```

```python
# Calcular el promedio de calificaciones por materia
promedio_por_materia = df_calificaciones.groupby('Materia')['Calificación'].mean()
print(promedio_por_materia)

# Calcular el promedio de calificaciones por estudiante
promedio_por_estudiante = df_calificaciones.groupby('Estudiante')['Calificación'].mean()

# Encontrar el estudiante con el promedio más alto
estudiante_promedio_max = promedio_por_estudiante.idxmax()
print(f"El estudiante con el promedio más alto es: {estudiante_promedio_max}")

# Calcular el promedio de calificaciones por estudiante
promedio_por_estudiante = df_calificaciones.groupby('Estudiante')['Calificación'].mean()
print(promedio_por_estudiante)

# Contar estudiantes con promedio superior a 85
estudiantes_promedio_alto = promedio_por_estudiante[promedio_por_estudiante > 85].count()
print(f"Estudiantes con promedio superior a 85: {estudiantes_promedio_alto}")

# Encontrar la materia con la mayor cantidad de calificaciones
materia_mas_calificaciones = df_calificaciones['Materia'].value_counts().idxmax()
print(f"La materia con la mayor cantidad de calificaciones es: {materia_mas_calificaciones}")

# Mostrar los 5 estudiantes con el promedio más bajo
estudiantes_promedio_bajo = promedio_por_estudiante.sort_values().head(5)
print("Los 5 estudiantes con el promedio más bajo son:")
print(estudiantes_promedio_bajo)
```
